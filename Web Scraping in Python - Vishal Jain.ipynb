```python
from bs4 import BeautifulSoup #Import the BeautifulSoup module from bs4 as convention.
```

```python
import requests
```

```python
import pandas as pd
from pandas import Series,DataFrame  # Import the Series and DataFrame from pandas
import numpy as np
```

```python
#url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
url = 'https://en.wikipedia.org/wiki/List_of_current_Indian_chief_ministers'
```

```python
# Setup a request and grab the content as a BeautifulSoap object.
result = requests.get(url)
c = result.content
```

```python
soup = BeautifulSoup(c,'lxml') # Beautifulsoup takes in two argument (HTML string to be parsed , optional : HTML Parser name)
```

```python
type(soup) # The Soup Object
```

```python
soup.text # The BeautifulSoup object has a text attribute that returns the plain text of a HTML string sans the tags.
```

```python
summary = soup.find_all('table',{"class" : "wikitable sortable"})
```

```python
summary
```

```python
# Jump to area of Interest on the webpage. This needs basic HTML knowledge. Right click on HTML page and click inspect option. Search the relevant tag there.
summary = soup.find_all('table',{"class" : "wikitable sortable"}) # This brings only the desired table data in text form.
# summary = soup.find_all('table') # This is not used because it will bring first table with color key for parties data also.
```

```python
rows = summary[0].findAll('tr')
```

```python
data = [] # Creation of Empty List which will store data.
```

```python
#Extract Table Row and Table Data information from HTML tag. This is 
for tr in rows:
    cols =  tr.findAll('td')
    for td in cols:
        text = td.find(text = True)
        print (text)
        data.append(text)
    
```

```python
type(data)
```

```python
for text in data:  #To remove \n character from the list.
    if '\n' in data:
        data.remove('\n')
```

```python
data
```

```python
len(data) # 155 elements are present in the list.
```

```python
# Below code is used to separate the data into relevant 5 lists. This will be used to create Pandas DataFrame. DataFrame will be used for any further data analysis in an easy way.
state = []
cm = []
doj = []
party = []
num = []
i=0
for _ in data:
    for _ in data:
        if len(data)>=i+5:
            state.append(data[i])
            cm.append(data[i+1])
            doj.append(data[i+2])
            party.append(data[i+3])
            num.append(data[i+4])
            i+=5
        


```

```python
for element in zip(cm,state,doj,party): # This just to see how data looks when 4 lists are zipped together.
    print(element)
```

```python
# Creation of DataFrame using list.
dataframe = pd.DataFrame({'Chief Minister' : cm , 'State' : state , 'Date' : doj , 'Political Party' : party  }) 
```
